```python
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
x = np.linspace(0.8, 10, 100, endpoint=True)
fig, ax = plt.subplots()
ax.plot(x,x/np.exp(x),label='x/ex')
ax.plot(x,pow(x,2)/np.exp(x),label='x^2/ex')
ax.plot(x,pow(x,3)/np.exp(x),label='x^3/ex')
ax.plot(x,pow(x,4)/np.exp(x),label='x^4/ex')
ax.plot(x,pow(x,5)/np.exp(x),"--b",label='x^5/ex')
leg = ax.legend();
ax.legend(loc='upper right', frameon=False)
```

```python
fig, ax = plt.subplots()
ax.plot(x,np.log(x)/x,label='ln(x)/x')
ax.plot(x,np.log(x)/pow(x,2),label='ln(x)/x^2')
ax.plot(x,np.log(x)/pow(x,3),label='ln(x)/x^3')
leg = ax.legend();
ax.legend(loc='upper right', frameon=False)
```

```python
fig, ax = plt.subplots()
y = np.linspace(-10, 10, 100, endpoint=True)
ax.plot(y,np.exp(y)/(1+np.exp(y)))
```
