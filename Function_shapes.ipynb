```python
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
x = np.linspace(0.1, 5, 100, endpoint=True)
fig, ax = plt.subplots()
ax.plot(x,np.exp(x),label='e^x')
leg = ax.legend();
ax.legend(loc='upper right', frameon=False)
```

```python
fig, ax = plt.subplots()
ax.plot(x,np.log(x),label='ln(x)')
leg = ax.legend();
ax.legend(loc='upper right', frameon=False)
```
