# TL of Expansion Chamber Silencer

## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Iniatialized vector

```python
x = np.linspace(50,2000,500)
```

## Input parameters
d = inlet/outlet duct diameter (m)
D = silencer diameter (m)
l = length of silencer (m)
c = speed of sound (m/s)

```python
d = 0.05
D = 0.15
m = (D/d)**2
l = 0.3
c = 343
TL = 10*np.log10(1+0.25*(m-(1/m))**2*np.sin(2*np.pi*x*l/c)**2)
plt.plot(x,TL)
plt.show()
```
